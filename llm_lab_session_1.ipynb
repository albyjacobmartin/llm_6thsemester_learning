```python
import nltk
from nltk.tokenize import word_tokenize

sentence = input()

tokens=word_tokenize(sentence)
print(f"The tokenized words are: {tokens}")
```

```python
doc = nlp(sentence)
word_embeddings = []

for token in doc:
    if token.has_vector:
        word_embeddings.append(token.vector)

print(f"Number of word embeddings generated: {len(word_embeddings)}")
if word_embeddings:
    print(f"Shape of the first word embedding: {word_embeddings[0].shape}")
else:
    print("No word embeddings were generated.")
```
